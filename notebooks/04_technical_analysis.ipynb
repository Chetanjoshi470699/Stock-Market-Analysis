# 04 – Technical Analysis

Computes and visualises technical indicators:
- RSI (14-day)
- Bollinger Bands (20-day, 2σ)
- MACD (12/26/9)
- ATR (14-day)
- Stochastic Oscillator (%K / %D)
- On-Balance Volume (OBV)
- Interactive candlestick chart (Plotly)

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import warnings; warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from src.preprocessing import load_processed
from src.analysis import TechnicalAnalysis
from src.visualization import StockVisualizer

TICKER = 'AAPL'
df = load_processed(TICKER)
ta = TechnicalAnalysis(df)
df = ta.add_all_indicators()

print(f'Loaded {TICKER}: {df.shape}')
df.tail(3)
```

## 4.1 Candlestick Chart (Interactive – Plotly)

```python
viz = StockVisualizer(df, ticker=TICKER)
viz.plot_candlestick(show_bb=True, show_volume=True, last_n=252, save=True)
```

## 4.2 RSI (14-day)

```python
viz.plot_rsi(last_n=252, save=True)
```

### RSI Signal Audit

```python
oversold  = df[df['rsi'] < 30][['close', 'rsi']].tail(10)
overbought = df[df['rsi'] > 70][['close', 'rsi']].tail(10)

print('--- Oversold signals (RSI < 30) ---')
print(oversold.to_string())
print('\n--- Overbought signals (RSI > 70) ---')
print(overbought.to_string())
```

## 4.3 Bollinger Bands

```python
plot_df = df.tail(252)

fig, axes = plt.subplots(2, 1, figsize=(14, 9), sharex=True)

# Price + bands
axes[0].plot(plot_df.index, plot_df['close'],    label='Close',    color='#ecf0f1', linewidth=1.2)
axes[0].plot(plot_df.index, plot_df['bb_upper'], label='BB Upper', color='#3498db', linestyle='--', linewidth=1)
axes[0].plot(plot_df.index, plot_df['bb_mid'],   label='BB Mid',   color='#f1c40f', linestyle='--', linewidth=1)
axes[0].plot(plot_df.index, plot_df['bb_lower'], label='BB Lower', color='#3498db', linestyle='--', linewidth=1)
axes[0].fill_between(plot_df.index, plot_df['bb_upper'], plot_df['bb_lower'], alpha=0.07, color='#3498db')
axes[0].set_title(f'{TICKER} – Bollinger Bands', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Price (USD)')
axes[0].legend(loc='upper left', fontsize=8)

# %B
axes[1].plot(plot_df.index, plot_df['bb_pct'], color='#e74c3c', linewidth=1)
axes[1].axhline(1.0, color='#e74c3c', linestyle='--', alpha=0.6)
axes[1].axhline(0.0, color='#26a69a', linestyle='--', alpha=0.6)
axes[1].axhline(0.5, color='white',   linestyle='--', alpha=0.3)
axes[1].set_ylabel('%B')
axes[1].set_ylim(-0.2, 1.4)

plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('../images/charts/AAPL_bollinger_bands.png', dpi=150)
plt.show()
```

## 4.4 MACD

```python
viz.plot_macd(last_n=252, save=True)
```

## 4.5 ATR – Measuring Market Volatility

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

axes[0].plot(df.index, df['close'], color='#ecf0f1', linewidth=1.2)
axes[0].set_title(f'{TICKER} – Price & ATR', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Price (USD)')

axes[1].plot(df.index, df['atr'], color='#f39c12', linewidth=1.2)
axes[1].set_ylabel('ATR (14-day)')
axes[1].fill_between(df.index, df['atr'], alpha=0.25, color='#f39c12')

plt.tight_layout()
plt.savefig('../images/charts/AAPL_atr.png', dpi=150)
plt.show()
```

## 4.6 Stochastic Oscillator

```python
plot_df = df.tail(252)

fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

axes[0].plot(plot_df.index, plot_df['close'], color='#ecf0f1', linewidth=1.2)
axes[0].set_title(f'{TICKER} – Stochastic Oscillator', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Price (USD)')

axes[1].plot(plot_df.index, plot_df['stoch_k'], label='%K', color='#3498db', linewidth=1.2)
axes[1].plot(plot_df.index, plot_df['stoch_d'], label='%D', color='#e74c3c', linewidth=1.2, linestyle='--')
axes[1].axhline(80, color='red',   linestyle=':', alpha=0.6)
axes[1].axhline(20, color='green', linestyle=':', alpha=0.6)
axes[1].set_ylim(0, 100)
axes[1].set_ylabel('Stochastic')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.savefig('../images/charts/AAPL_stochastic.png', dpi=150)
plt.show()
```

## 4.7 On-Balance Volume (OBV)

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

axes[0].plot(df.index, df['close'], color='#ecf0f1', linewidth=1.2)
axes[0].set_title(f'{TICKER} – Price & OBV', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Price (USD)')

axes[1].plot(df.index, df['obv'] / 1e9, color='#2ecc71', linewidth=1.2)
axes[1].set_ylabel('OBV (B)')

plt.tight_layout()
plt.savefig('../images/charts/AAPL_obv.png', dpi=150)
plt.show()
```
